```python
import cv2
import numpy
from matplotlib import pyplot as plt
```

```python
def displayImage(image, title):
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
image = cv2.imread("./data/logo.png")
```

```python
print(image)
```

```python
print("image shape %s" % str(image.shape))
```

```python
print("image width: %d " % image.shape[0])
```

```python
print("image height: %d " % image.shape[1])
```

```python
print("image channels: %d " % image.shape[2])
```

```python
print("(b g r) = %s" % str(image[200,100]))
```

```python
plt.imshow(image)
plt.title('logo')
plt.show()
```

```python
cv2.imwrite("logo.jpg", image)
```

```python
w = image.shape[0]
h = image.shape[1]
(cx, cy) = (w//2, h//2)
print(cx)
print(cy)
```

```python
top_left = image[0:cy,0:cx]
#top_left = (0,255,0)
top_right = image[0:cy, cx:w]
bottom_left = image[cy:h, 0:cx]
bottom_right = image[cy:h, cx:w]

displayImage(top_left, "top_left")
displayImage(top_right, "top_right")
displayImage(bottom_left, "bottom_left")
displayImage(bottom_right, "bottom_right")















```
